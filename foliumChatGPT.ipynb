```python
import folium

# Coordenadas das cidades
coordenadas = {
    'Porto Alegre, RS, Brasil': [-30.0346, -51.2177],
    'São Paulo, SP, Brasil': [-23.5505, -46.6333],
    'Rio de Janeiro, RJ, Brasil': [-22.9068, -43.1729],
    'Curitiba, PR, Brasil': [-25.4290, -49.2671],
    'Belo Horizonte, MG, Brasil': [-19.9173, -43.9345]
}

# Cria o mapa centrado em São Paulo
mapa = folium.Map(location=coorden
adas['São Paulo, SP, Brasil'], zoom_start=5)

# Dicionário de cores para os marcadores
cores = {
    'Porto Alegre, RS, Brasil': 'lightblue',
    'São Paulo, SP, Brasil': 'purple',
    'Rio de Janeiro, RJ, Brasil': 'red',
    'Curitiba, PR, Brasil': 'black',
    'Belo Horizonte, MG, Brasil': 'green'
}

# Adiciona os marcadores no mapa
for cidade, coord in coordenadas.items():
    folium.Marker(
        location=coord,
        popup=cidade,
        icon=folium.Icon(color=cores[cidade], icon_color='white')
    ).add_to(mapa)

# Exibe o mapa
mapa
```
