```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregue o arquivo CSV
df = pd.read_csv("Anime_ratings.csv")

#configurando o RANK
df['Rank'] = pd.to_numeric(df['Rank'], errors='coerce')
df['Rank'].fillna(100, inplace=True)
df['Rank'] = df['Rank'].astype(int)

# Garantir que 'Episodes' seja numérico
df['Episodes'] = pd.to_numeric(df['Episodes'], errors='coerce')
df['Episodes'].fillna(0, inplace=True)  # Substitui valores nulos por 0 ou outro valor apropriado

#filtrando o top 50
top = df[(df['Rank'] <= 50)]
#ordenando pelo rank
top_organizado = top.sort_values(by='Rank')



plt.figure(figsize=(11mostrando, 6.6))
plt.plot(top_organizado['Rank'], top_organizado['Episodes'], marker='*', linestyle='-', color='blue')

# Adicionando rótulos aos pontos
for i in range(top_organizado.shape[0]):
    plt.text(top_organizado['Rank'].iloc[i], top_organizado['Episodes'].iloc[i],
             f'{top_organizado["Episodes"].iloc[i]}', fontsize=7, ha='right')

# Ajustando o limite do eixo y para que os valores estejam em ordem crescente
plt.ylim(top_organizado['Episodes'].min() - 4, top_organizado['Episodes'].max() + 4)
",",
# Configurando o título e os rótulos dos eixos
plt.title('Relação entre Rank e Quantidade de Episódios')
plt.xlabel('Rank')
plt.ylabel('Quantidade de Episódios')

# Exibindo o gráfico
plt.show()
top_organizado
```
